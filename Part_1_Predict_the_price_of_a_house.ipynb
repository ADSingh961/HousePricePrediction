Input Split

```python
X=data.iloc[:,:-1]
y=data['Total Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

Model Training

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
# Score for Linear Regression Model
lr.score(X_train,y_train)
```

```python
# Introducing various other Models
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
knn=KNeighborsRegressor()
dt=DecisionTreeRegressor()
rf=RandomForestRegressor()
```

```python
from sklearn.metrics import accuracy_score
```

```python
def func(model):
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    print(r2_score(y_test,y_pred))
```

```python
# Score for K_Nearest Neighbour
func(knn)
```

```python
# Score for Decision Tree Regressor
func(dt)
```

```python
# Score for Random Forest Regressor
func(rf)
```

**Linear Regressor Model has the highest score against all other models**
